**Introduction to Matplotlib and Pyplot**


Matplotlib is a python plotting library which produces publication quality figures.Pyplot is a collection of functions that make matplotlib work like Matlab.Pyplot is used in Data Visualizations.
Pyplot provides what is called state machine interface to matplot library.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Plt plot function can be used to plot lines and markers.In its most basic form we need only one argument for the command and it plots the argument as y value corresponding to indexes as the x values.

```python
plt.plot([0,1,4,9,16]) #Python returns a matplotlib object if we do not add a semi colon in the end of the command.
```

```python
plt.plot([0,1,4,9,16]);
```

If provided with two arguments then the first argument deals with the x values and second corresponds to the y values for the plot

```python
x=np.linspace(0,10,20)
```

```python
y=x**2
```

```python
plt.plot(x,y)
```

Third argument to this function is the format string which specifies the color, marker, and line type.

```python
y2=x**1.5
```

```python
plt.plot(x,y,'bo-',linewidth=2,markersize=5);# b:blue,o:markers for the points,- for the line
```

```python
plt.plot(x,y,'gs-',linewidth=2,markersize=5);
```

**Customizing your Plots**

We can add legend using legend function, adjust axis with axis, set axis labels using x labels and y labels and save figures using save figs where file format specifies which format to save it in.
Latex is also supported by plt,to use it use $ signs around the symbol.


```python
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y,'bd-')
plt.plot(x,y2,'go-')
plt.axis([-0.5,12,0,140])
```

plt.axis([xmin,xmax,ymin,ymax]) is the format for axis values. We can also add labels and legends to the plot as follows:


```python
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x,y,'bd-',label='First')
plt.plot(x,y2,'go-',label='Second')
plt.axis([-0.5,12,0,140])
plt.legend(loc= 'upper right')
plt.savefig('myplot.pdf')
```
