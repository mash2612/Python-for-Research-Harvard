We use np.random.normal() function to generate standard normal distribution which has a mean of 0 variance =1 and values between 0 and 1 .

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.normal(size=1000)#Generating 1000 numbers from standard normal distribution
```

```python
plt.hist(x);#Note we have number of observations on y-axis
```

By default hist uses 10 evenly spaced bins and it tries to optimize bin width and bin location.

```python
plt.hist(x,normed=True)# We have proportion of values falling in each bin
```

Specifying the location in this case of the bins using location function

```python
plt.hist(x,normed=True,bins=np.linspace(-5,5,20));
```

Considering Gamma Distribution now which starts at 0 and goes all the way to positive infinity.Like normal distribution this also has two parameters.We also use a function called subplot which allows us to have many subplots simultaneously in one figure.There are three arguments to this function,first two are rows and columns and the third is the plot number.Plot number always starts at 1.They are incremented across row first.

```python
x=np.random.gamma(2,3,10000)
```

```python
plt.hist(x, bins=30);
```

```python
plt.hist(x, bins=30, normed=True);
```

```python
plt.hist(x, bins=30, cumulative=True, normed=True);
```

```python
plt.hist(x, bins=30, cumulative=True, normed=True, histtype='step');
```

We can create all the histograms in one figure as subplots

```python
plt.figure()
plt.subplot()
plt.subplot(221)
plt.hist(x, bins=30);
plt.subplot(222)
plt.hist(x, bins=30, normed=True);
plt.subplot(223)
plt.hist(x, bins=30, cumulative=True, normed=True, histtype='step');
plt.subplot(224)
plt.hist(x, bins=30, cumulative=True, normed=True);
```
