In some plots we need x and y coordinates to be transformed using log function.So the value plotted on x is not the original x axis value rather the log value of x and similarly for y.
Log by default is base 10.
semilogx() plots x on log and y on original axis and vice versa for semilogy().
loglog plots both x and y on log axis.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,20)
```

```python
y=x**2
```

```python
y1=x**1.5
```

```python
plt.loglog(x,y,'bo-',label='First');
plt.loglog(x,y1,'gd-',label='Second');
plt.legend(loc='upper right');
```

Notice here that the spacing is not even as the spacing is linear and axis are log.To accomplish even spacing on this plot we can use np.logspace()

```python
x=np.logspace(-1,1,40)#Corresponding to x=0.1 to x=10 with 40 points in between
y=x**2
y1=x**1.5
plt.loglog(x,y,'bo-',label='First');
plt.loglog(x,y1,'gd-',label='Second');
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc='upper right');
```
